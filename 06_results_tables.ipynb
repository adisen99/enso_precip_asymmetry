```python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy import stats
import os
```

```python
def get_model_name_runs(loc):
    list_dfs = []
    for file in sorted(os.listdir(loc)):
        if '.csv' in file:
            df = pd.read_csv(loc + '/' + file)[['source_id', 'variant_label', 'variable_id', 'nominal_resolution']].iloc[0]
            list_dfs.append(df)
    out = pd.concat(list_dfs, axis=1)
    return out
```

```python
pd.read_csv('./data/model/ACCESS-ESM1-5_pr.csv')[['source_id', 'variant_label', 'variable_id', 'nominal_resolution']].sort_values('variant_label')#.iloc[0]
```

```python
out_df = get_model_name_runs('./data/model')
```

```python
out_df
```

```python
table = pd.DataFrame(data=dict(model = out_df.iloc[0, :], run = out_df.iloc[1, :], variable = out_df.iloc[2,:], res = out_df.iloc[3,:]))
table
```

```python
def agg_func(x):
    return '; '.join(x)
```

```python
agg_functions = dict(run = agg_func, variable = agg_func, res = agg_func)
```

```python
table_final = table.groupby(table['model']).aggregate(agg_functions)
table_final
```

```python
table_final.shape
```

```python
table_final.to_csv('./data/result/model_table.csv')
```
